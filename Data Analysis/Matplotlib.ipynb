# Matplotlib

[Matplotlib](https://matplotlib.org/) is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used in data science, machine learning, and scientific computing to create a variety of plots and graphs, including line plots, bar charts, scatter plots, histograms, and more.

## Key Features
- **Customizable Plots**: Create highly customizable visualizations with control over every element.
- **Wide Range of Charts**: Supports a variety of chart types, from simple plots to complex multi-figure layouts.
- **Integration**: Works seamlessly with libraries like NumPy, pandas, and Jupyter Notebooks.
- **Output Formats**: Supports multiple output formats, such as PNG, PDF, SVG, and interactive backends.

````python
import matplotlib.pyplot as plt 
````

````python
x = [1,2,3,4,5]
y = [2,3,4,5,10]
plt.plot(x,y) # this is for the line plot 
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Simple Line Plot")
plt.show()
````

````python
x = [1,2,3,4,5]
y = [2,5,6,3,10]
plt.plot(x,y, color = 'blue' , linestyle = '--' , marker = 'o' , linewidth = 1.5) # This is how we can able to style the plot accordingly
plt.grid(True)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Line Plot with Different Color and Style")
plt.show()
````

````python
x = [1,2,3,4,5]
y = [2,3,4,5,6]
z = [5,2,6,7,3]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y,color = 'red' , linestyle = '--' , marker = 'o')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y,x, color = 'green' , linestyle='-.', marker = 'o')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,z ,color = 'blue' , linestyle='-.', marker = 'o')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(z,y,color = 'brown' , linestyle='-.', marker = 'o')
plt.title("Plot 3")
plt.show()
````
